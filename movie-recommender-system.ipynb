```python
# importing the libraries
import numpy as np
import pandas as pd
```

```python
# import the data set that we need for our project 
movies= pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
# Here we have to merge the two data set movies and credits on the basis of title


# Assuming you have two DataFrames: movies and credits
# Merge the DataFrames on the 'title' column
movies = movies.merge(credits, on='title')

# Get the shape of the merged DataFrame
shape_of_merged_movies = movies.shape

# Display the shape
print(shape_of_merged_movies)
```

```python
# The thing we have to keep for our project 
# genres
# id
# keywords
# title
# overview
# cast
# crew
movies = movies[['movie_id' , 'title' , 'overview' , 'genres' , 'keywords' , 'cast' , 'crew']]
```

```python
# main data on which we work for our recommender system
movies.head()
```

```python
movies.dropna(inplace= True)
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name']) 
    return L 
        
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
# we have to convert the data in the format of 
#   ['Action', 'Adventure' , 'Fantasy' , 'Science Fiction']
```

```python
movies.head()
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# In the cast we have to need the top three cast only so we have to modify the funtion
def convert3(obj):
    L=[]
    counter =0
    for i in ast.literal_eval(obj):
         if counter !=3:
            L.append(i['name'])
            counter+=1
         else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(obj):
    L =[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['overview']
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x:[i.replace("  ", "") for i in x])
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace("  ","") for i in x])
movies['cast']= movies['cast'].apply(lambda x: [i.replace("  ","") for i in x])
movies['crew']= movies['crew'].apply(lambda x: [i.replace("  ","") for i in x])
```

```python
movies.head()
```

```python
# we have to create a new column so the we can concat the 'keyword' , 'cast' , 'crew' , 'overview' , 'genres'
movies['tags']= movies['overview']+ movies['genres']+ movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
# now we have to form the new data frame so that we can only need the tag column
new_df = movies[[ 'movie_id' , 'title' ,'tags']]
new_df
```

```python
# Now we have to convert the ['loved' , 'loving' , 'love'] to ['love' , 'love' , ' love']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)    
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
CV = CountVectorizer(max_features= 5000 , stop_words= 'english')
```

```python
vectors = CV.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
CV.get_feature_names_out()
```

```python
# now every movie is converted in a vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
cosine_similarity(vectors).shape
```

```python
similarity
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie ].index[0]
    distances = similarity[movie_index]
    movie_list = sorted(list(enumerate(distances)) , reverse=True , key = lambda x : x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
    return
    
```
